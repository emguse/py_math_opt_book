Let’s go through step by step to build a mathematical optimization model to maximize the amount of cookie dough and cake dough that can be produced given the available flour and butter.

Step 1: Define the decision variables

The first step is to define the decision variables, which represent what we are trying to optimize.

- Let x represent the amount of cookie dough produced (in grams).
- Let y represent the amount of cake dough produced (in grams).

Both x and y are non-negative real numbers.

Step 2: Write the objective function

Our goal is to maximize the total amount of dough produced. So the objective function is to maximize x + y, where:

- x is the cookie dough in grams.
- y is the cake dough in grams.

Objective Function:

Maximize x + y

Step 3: Identify the constraints

Next, we consider the limitations or constraints, which are based on the available resources (flour and butter).

Flour constraint:
- Each 10g of cookie dough requires 3g of flour.
- Each 10g of cake dough requires 4g of flour.
- The total amount of flour available is 1200g.
The constraint for flour is:

0.3x + 0.4y <= 1200

Butter constraint:
- Each 10g of cookie dough requires 3g of butter.
- Each 10g of cake dough requires 2g of butter.
- The total amount of butter available is 900g.
The constraint for butter is:

0.3x + 0.2y <= 900

Step 4: Non-negativity constraints

Both x and y represent the amounts of dough produced, and they cannot be negative. Thus, we add the non-negativity constraints:

x >= 0, y >= 0

Step 5: Complete mathematical model

Now, combining the objective function and all constraints, the mathematical model for maximizing cookie and cake dough production is:

Model:

Maximize x + y

Subject to:

0.3x + 0.4y <= 1200 (Flour constraint)

0.3x + 0.2y <= 900 (Butter constraint)

x >= 0, y >= 0 (Non-negatibity constraints)
